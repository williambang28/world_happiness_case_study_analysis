```python
#import libraries
import pandas as pd
```

```python
#clean data

df = pd.read_csv("data.csv") # define dataframe

df.columns = df.columns.str.lower().str.replace(" ", "_").str.replace("explained_by:_", "") # standardise column names

df = df.dropna(axis="columns", thresh=1013) # remove columns with minimum 1013 (2116 - 1103) non-null values
df = df.drop(columns=["upper_whisker", "lower_whisker"]) # remove life_eval lq and uq

df = df.rename (columns= {
    "life_evaluation_(3-year_average)": "life_eval", # rename columns
    "country_name": "country",
    "log_gdp_per_capita": "gdp_per_capita",
    "freedom_to_make_life_choices": "freedom",
    "perceptions_of_corruption": "corruption",
    "dystopia_+_residual": "dystopia_residual"
})

df.to_csv("cleaned_data.csv", index=False) # save to new file

# source: https://www.worldhappiness.report/data-sharing/
```
